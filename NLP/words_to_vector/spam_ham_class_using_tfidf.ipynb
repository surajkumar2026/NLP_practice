```python

messages = pd.read_csv(r"e:\NLP_practice\Data\spam.csv",   usecols=[0, 1],skiprows=1,  # Read only the first two columns
    names=["label", "message"],encoding="ISO-8859-1")
print(messages)
```

```python
##data cleaning 
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
WordNetLemmatize= WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages["message"][i])
    review = review.lower()
    review = review.split()
    review = [WordNetLemmatize.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## output fetaure
y = pd.get_dummies(messages['label'])
```

```python
y
```

```python
y=y.iloc[:,0].values
```

```python
#train test split 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(corpus,y,test_size=0.20)
```

```python
len(X_train),len(y_train)
```

```python
#creat th ebag of wordsd
from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
#this my independent feature
X_train_vec = tv.fit_transform(X_train).toarray()
X_trest_vec=tv.transform(X_test).toarray()

```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model= MultinomialNB()
```

```python
spam_detect_model.fit(X_train_vec,y_train)
```

```python
y_pred=spam_detect_model.predict(X_trest_vec)
```

```python
#check perfromance matrix
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
